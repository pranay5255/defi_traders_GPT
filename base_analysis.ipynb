```python
df = pd.DataFrame()
for file in file_list:
    filename = os.path.join("VCwallettxns",file)
    temp_df = pd.read_json(filename)
    df = pd.concat([df,temp_df])

    

```

```python
df.reset_index(inplace=True)
```

```python
vc_names =  pd.read_csv("VC_wallets.csv",index_col=[0])
```
